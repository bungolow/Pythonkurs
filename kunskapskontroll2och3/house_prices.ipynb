```python

# Kontrollera värden i ocean_proximity kolumnen
print(housing["ocean_proximity"].unique())
print("")

# Summan av alla hushåll indelade i ocean_proximity
households_by_ocean = (
    housing
    .groupby("ocean_proximity")["households"]
    .sum()
    .sort_values(ascending=False)
)

print(households_by_ocean)

# Plottar
plt.figure(figsize=(9, 5))
plt.bar(
    households_by_ocean.index,
    households_by_ocean.values,
    edgecolor="black"
)

plt.xlabel("Ocean Proximity")
plt.ylabel("Total Households")
plt.title("Total Households by Ocean Proximity")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python

california_img = mpimg.imread('california.png')

# Create a color for each category
categories = housing['ocean_proximity'].unique()
colors = plt.cm.tab10(np.linspace(0, 1, len(categories)))  # categorical colormap
color_dict = dict(zip(categories, colors))

# Plot the background map first
plt.figure(figsize=(12,8))
plt.imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5, zorder=0)

# Plot each point
for category in categories:
    subset = housing[housing['ocean_proximity'] == category]
    plt.scatter(
        subset['longitude'], 
        subset['latitude'], 
        s=subset['population']/100,   # size proportional to population
        color=color_dict[category],
        alpha=0.6,
        label=category,
        edgecolor='k',
        linewidth=0.3
    )

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("California Housing Population by Ocean Proximity", fontsize=16)
plt.legend(title="Ocean Proximity")
plt.show()
```
